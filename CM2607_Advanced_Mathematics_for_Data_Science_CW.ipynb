# Problem Statement

## 1. Modelling Travel Time

**Let:**
*   $x$ = total route distance (in km)
*   $y$ = traffic delay factor (in minutes per km)

The travel time for each mode of transport is modelled as:

$$T_b(x, y) = ax^2 + by^2$$

$$T_t(x, y) = \sqrt{c(x^2 + y^2)}$$

**Where:**
*   $a$ reflects the effect of distance on bus travel time (stops, boarding times).
*   $b$ reflects the effect of congestion on bus travel time.
*   $c$ is a proportionality constant converting combined effects into minutes for taxis.

**(a) Calculate the gradient $\nabla T$ for each mode.**

> **Hint:** The gradient of a scalar field $T$ is defined as:
> $$\nabla T = \left( \frac{\partial T}{\partial x}, \frac{\partial T}{\partial y} \right)$$

**(b) Interpret which factor has a greater impact for bus and taxi travel times.**

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import sympy as sp
```

```python
x, y, z, a, b, c = sp.symbols('x y z a b c')
```

## Taxi Travel Time

```python
Tb = a*x**2 + b*y**2
```

```python
display(Tb)
```

```python
Tb_x = sp.diff(Tb, x)
display(Tb_x)
```

```python
Tb_y = sp.diff(Tb, y)
display(Tb_y)
```

```python
gradient_Tb = sp.Matrix([Tb_x, Tb_y])
display(gradient_Tb)
```

## Bus Trave Time

```python
Tt = sp.sqrt(c*(x**2 + y**2))
```

```python
display(Tt)
```

```python
Tt_x = sp.diff(Tt, x)
display(Tt_x)
```

```python
Tt_y = sp.diff(Tt, y)
display(Tt_y)
```

```python
gradient_Tt = sp.Matrix([Tt_x, Tt_y])
display(gradient_Tt)
```
